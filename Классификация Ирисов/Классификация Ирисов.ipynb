```python
import numpy as np
import pandas as
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
import mglearn
from pandas.plotting import scatter_matrix 
from sklearn.model_selection import train_test_split
%matplotlib inline 
sns.set(style="ticks")
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print(iris_dataset.keys())
```

```python
print(type(iris_dataset["data"]))
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset["data"],iris_dataset["target"], random_state = 0)
```

```python
# создаем dataframe из данных в массиве X_train
# маркируем столбцы, используя строки в iris_dataset.feature_names
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
# создаем матрицу рассеяния из dataframe, цвет точек задаем с помощью y_train
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
X_new.shape
```

```python
prediction = knn.predict(X_new)
print(prediction)
print("Спрогнозированная мета {}".format(iris_dataset["target_names"][prediction]))
```

```python
y_pred = knn.predict(X_test)
print(y_pred)
print("Правильность предугадывания тестового набора {:.2f}".format(np.mean(y_pred == y_test)))
print("Правильность предугадывания тестового набора {:.2f}".format(knn.score(X_test, y_test)))
```
